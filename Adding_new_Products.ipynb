###ADDING MODULE

1. ADMIN CAN ADD A NEW PRODUCT
2. ADMIN CAN ADD AN EXISTING PRODUCT
3. ADMIN CAN DELETE A PRODUCT

```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()

record = json.loads(r)
```

```python
record
```

####Add New or existing Item into Inventory

```python
prod_id = str(input("Enter product id:"))
category = str(input("Enter category:"))
name = str(input("Enter name:"))
pr = int(input("Enter price:"))
qn = int(input("Enter quantity:"))
discount = int(input("Enter discount in percentage:"))

for id in record:
  if id==prod_id:
    record[id]['qn']=record[id]['qn']+qn
    record[id]['pr']=pr
    record[id]['discount']=discount
    break
else:
  record[prod_id] = {'name': name, 'pr': pr, 'qn': qn, 'category':category, 'discount':discount, 'ratings':'Unrated','Times_purchased':0, 'times_rated': 0}

js = json.dumps(record)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```
